### 03) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10.1,20.2,30.3,40.4,50.5]
l=['1st','2nd','3rd','4th','5th']
c=['b','g','r','c','m']
w=[0.9,1,0.4,1,1]
plt.title("rank")
plt.bar(x,y,color=c,label=l,width=w)
plt.show()

```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=["Python","DAA","JS","Flutter"]
y=[100,25,9,65]
c=["r","g","m","c"]
plt.title("Popularity of languages",fontsize=15)
plt.xlabel("languages",fontsize=10,color='b')
plt.ylabel("%",fontsize=10,color='b')
plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.7)

def addlables(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
addlables(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=['b','g', 'r','y','m']
size=[100,200,300,400,500]
plt.scatter(x,y,color=c, s=size, alpha=0.9, marker="^", edgecolor='b')
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("x-axis", fontsize=15)
plt.ylabel("y-axis", fontsize=15)
plt.show()
```

### 03) WAP to Display an Image in Grayscale in Matplotlib.

```python
#Python Imaging Library
import PIL
import matplotlib.pyplot as plt
img =  PIL.Image.open("wood.jpg")
#1 (1-bit pixels, black and white, stored with one pixel per byte)
#L (8-bit pixels, black and white)
#P (8-bit pixels, mapped to any other mode using a color palette)
#RGB (3x8-bit pixels, true color)
#RGBA (4x8-bit pixels, true color with transparency mask)
#CMYK (4x8-bit pixels, color separation)
#YCbCr (3x8-bit pixels, color video format)
gray_img = img.convert("L")
plt.imshow(gray_img, cmap='gray')
```
